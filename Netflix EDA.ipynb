# Q. 2) In which year the highest number of TV Shows & Movies were released? Show with Bar Graph.

```python
df.head(6)
```

```python
release_count= df['release_year'].value_counts().sort_index()
release_count
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=release_count.index,y=release_count.values)
plt.xticks(rotation = 90)
plt.xlabel('Release Year')
plt.ylabel('Number of Release')
```

# Q. 3) How many Movies & TV Shows are in the dataset ? Show with Bar Graph.

```python
type_counts=df['type'].value_counts()
type_counts
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=type_counts.index,y=type_counts.values)
plt.xticks(rotation = 90)
plt.xlabel('number of tv shows and Movie')
plt.ylabel('TV Shows and Movies')
```
